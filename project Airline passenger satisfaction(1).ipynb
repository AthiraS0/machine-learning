# OUTLIER REMOVAL

```python
def iqr_capping(df,variables):
    q1=df[variables].quantile(0.25)
    q3=df[variables].quantile(0.75)
    iqr=q3-q1
    upper_whisker=q3+(1.5*iqr)
    lower_whisker=q1-(1.5*iqr)
    return lower_whisker,upper_whisker
```

```python
lower_lim,upper_lim=iqr_capping(data,'Flight Distance')
print('lower lim = ',lower_lim)
print('upper lim = ',upper_lim)
```

```python
data['Flight Distance']=np.where(data['Flight Distance']>upper_lim,upper_lim,
                    np.where(data['Flight Distance']<lower_lim,lower_lim,data['Flight Distance']))
```

```python
lower_lim,upper_lim=iqr_capping(data,'Checkin service')
print('lower lim = ',lower_lim)
print('upper lim = ',upper_lim)
data['Checkin service']=np.where(data['Checkin service']>upper_lim,upper_lim,
                    np.where(data['Checkin service']<lower_lim,lower_lim,data['Checkin service']))
```

```python
lower_lim,upper_lim=iqr_capping(data,'Departure Delay in Minutes')
print('lower lim = ',lower_lim)
print('upper lim = ',upper_lim)
data['Departure Delay in Minutes']=np.where(data['Departure Delay in Minutes']>upper_lim,upper_lim,
                    np.where(data['Departure Delay in Minutes']<lower_lim,lower_lim,data['Departure Delay in Minutes']))
```

# SCALING

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2, random_state=32)
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_train=sc.fit_transform(x_train)
x_test=sc.fit_transform(x_test)
```

# Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
logit_model=LogisticRegression()
logit_model.fit(x_train,y_train)
y_pred=logit_model.predict(x_test)
```

```python
y_pred
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score
cm=confusion_matrix(y_test,y_pred)
print('Confusion matrix of logistic regression is',cm)
print('Accuracy of Logistic Regression is ',accuracy_score(y_test,y_pred))
```

```python
accuracy_lr=accuracy_score(y_test,y_pred)
print(accuracy_lr)
```

```python
from sklearn.metrics import ConfusionMatrixDisplay
disp=ConfusionMatrixDisplay(cm)
disp.plot()
plt.show
```

# KNN

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
metric_k=[]
neighbors=np.arange(3,10)
```

```python
for k in neighbors:
    classifier = KNeighborsClassifier(n_neighbors=k,metric='minkowski',p=2)
    classifier.fit(x_train,y_train)
    y_pred=classifier.predict(x_test)
    acc=accuracy_score(y_test,y_pred)
    metric_k.append(acc)
```

```python
plt.plot(neighbors,metric_k,'o-') 
plt.xlabel('k value')
plt.ylabel('accuracy')
plt.grid()
```

```python
for k in neighbors:
    classifier = KNeighborsClassifier(n_neighbors=3,metric='minkowski',p=2)
    classifier.fit(x_train,y_train)
    y_pred=classifier.predict(x_test)
    acc=accuracy_score(y_test,y_pred)
    metric_k.append(acc)
```
